# MNIST

Loading the data and reshaping it to work as input to the models

```python
from keras.datasets import mnist
from sklearn.model_selection import train_test_split

(X_train, y_train), (X_test, y_test)= mnist.load_data()
X_train_sm, X_test_sm, y_train_sm, y_test_sm = train_test_split(X_train, y_train, test_size=0.8, random_state=42)
X_train = X_train_sm.reshape(len(X_train_sm), 28*28)
X_test = X_test.reshape(len(X_test), 28*28)
y_train = y_train_sm
```

Grid search with cross validation

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

rf = RandomForestClassifier(verbose=1, n_jobs=-1)

params = {
        'criterion': ['gini', 'entropy'],
        'max_depth': [5, 10, 100, 200, 500],
        'min_samples_leaf': [1, 2, 5, 10],
        'max_features': ["sqrt", "log2"]
        }

gs = GridSearchCV(estimator=rf,param_grid=params,cv=5,n_jobs=-1, verbose=True, scoring='accuracy')
gs.fit(X_train, y_train)
```

```python
print('Best Parameters:',gs.best_params_,end='\n\n')
print('Best Score:',gs.best_score_)
```

Evaluation with standard parameters:

```python
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.ensemble import RandomForestClassifier
import seaborn as sns
import matplotlib.pyplot as plt

#Fitting and evaluating model 5 times with different random seeds(by not setting a specific random seed)
for i in range(5):
    print(f'Run {i+1}')
    rf = RandomForestClassifier(verbose=1, n_jobs=-1)
    rf.fit(X_train,y_train)

    pred = rf.predict(X_test)
    print(classification_report(y_test, pred))

    cf_matrix = confusion_matrix(y_test, pred)
    sns.set(rc={'figure.figsize':(11.7,8.27)})
    ax = sns.heatmap(cf_matrix/cf_matrix.astype(float).sum(axis=1), annot=True, fmt='.2%')
    ax.set_title('Random Forest')
    plt.show()
```

Averages 95% accuracy and f1-score.

Evaluation with optimized parameters:

```python
#Fitting and evaluating model 5 times with different random seeds(by not setting a specific random seed)
for i in range(5):
    print(f'Run {i+1}')
    rf = RandomForestClassifier(random_state=42, criterion='gini', max_depth=100, max_features='sqrt', min_samples_leaf=1, verbose=1, n_jobs=-1)
    rf.fit(X_train,y_train)

    pred = rf.predict(X_test)
    print(classification_report(y_test, pred))

    cf_matrix = confusion_matrix(y_test, pred)
    sns.set(rc={'figure.figsize':(11.7,8.27)})
    ax = sns.heatmap(cf_matrix/cf_matrix.astype(float).sum(axis=1), annot=True, fmt='.2%')
    ax.set_title('Random Forest')
    plt.show()
```

Avarages 95% accuracy and f1-score, same as with default parameters.

#  CIFAR-10

```python
from sklearn.model_selection import train_test_split
from keras.datasets import cifar10

(X_train, y_train), (X_test, y_test) = cifar10.load_data()
X_train_sm, X_test_sm, y_train_sm, y_test_sm = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
X_train = X_train_sm.reshape(len(X_train_sm), 3*32*32)
X_test = X_test.reshape(len(X_test_sm), 3*32*32)
y_train = y_train_sm
```

Grid search with cross validation.  We used less parameter to search over here since fitting the model to this dataset requires more compute.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

rf = RandomForestClassifier(random_state=42, verbose=1, n_jobs=-1)

params = {
        'max_depth': [100, 200],
        'min_samples_leaf': [1, 2, 5]
        }

gs = GridSearchCV(estimator=rf,param_grid=params,cv=5,n_jobs=-1, verbose=True, scoring='accuracy')
gs.fit(X_train, y_train)
```

```python
print('Best Parameters:',gs.best_params_,end='\n\n')
print('Best Score:',gs.best_score_)
```

Evaluation with standard parameters.

```python
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.ensemble import RandomForestClassifier

#Fitting and evaluating model 5 times with different random seeds(by not setting a specific random seed)
for i in range(5):
    print(f'Run {i+1}')
    rf = RandomForestClassifier(verbose=1, n_jobs=-1)
    rf.fit(X_train,y_train)

    pred = rf.predict(X_test)
    print(classification_report(y_test, pred))

    cf_matrix = confusion_matrix(y_test, pred)
    sns.set(rc={'figure.figsize':(11.7,8.27)})
    ax = sns.heatmap(cf_matrix/cf_matrix.astype(float).sum(axis=1), annot=True, fmt='.2%')
    ax.set_title('Random Forest')
    plt.show()
```

Average accuracy is 46%, f1-scores fluctuate a bit between 45-46%.

Evaluation with optimized parameters

```python
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns

#Fitting and evaluating model 5 times with different random seeds(by not setting a specific random seed)
for i in range(5):
    print(f'Run {i+1}')
    rf = RandomForestClassifier(max_depth=100, min_samples_leaf=2, verbose=1, n_jobs=-1)
    rf.fit(X_train,y_train)

    pred = rf.predict(X_test)
    print(classification_report(y_test, pred))

    cf_matrix = confusion_matrix(y_test, pred)
    sns.set(rc={'figure.figsize':(11.7,8.27)})
    ax = sns.heatmap(cf_matrix/cf_matrix.astype(float).sum(axis=1), annot=True, fmt='.2%')
    ax.set_title('Random Forest')
    plt.show()
```

Average accuracy and f1-score is basically 46%, although the first run gets 47% accuracy.

### Summary

For the simpler mnist dataset the random forest models does perform quite weel with an accuracy and f1-score of 95% on average. As is evident in our evaluation with MLPs and CNNs, there is room for improvement with more advanced model types. 

For the slightly more complex cifar-10 dataset, the performance is a lot worse at about 46% accuracy and f1-score. This shows that decision trees are not really fit for image classification as they do not take local relationships betweeen pixels into consideration. In image processing, each pixel can be seen as its own feature and therefore the data is quite high-dimensional. Decision trees work best if there are not too many features and if these features are meeningful in some way. As the pixels are treated as separate features, the decision trees will struggle to find correlations between them and the data would need to be pre-processed in some way to extract features from the images to be sent into the models during training. CNNs handle this process automatically and will therefore more or less always outperform decision trees in image classification tasks, altough this comes at the expense of being more complicated, less transparent(decision trees can be quite clearly visualized with all the conditions at each node) and more computationally expensive.
